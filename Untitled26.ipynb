```python
with open("xxx.txt", "r") as f:
    lines = [line.strip() for line in f.readlines() if line.strip()]

records = []
for i in range(0, len(lines), 2):
    part1 = lines[i]
    dataset = part1.split(",")[0].strip()
    score_model = part1.split("scores:")[1].split(",")[0].strip()
    expansion_model = part1.split("expansion:")[1].strip()

    part2 = lines[i + 1]
    time_value = float(part2.split("time:")[1].split(",")[0].strip())
    R_raw = part2.split("R =")[1].strip().strip("'")

    # Convert R to float (handle 'inf')
    R_value = float('inf') if R_raw == 'inf' else float(R_raw)

    records.append((dataset, score_model, expansion_model, time_value, R_value))

# for r in records:
#     print(f"Dataset: {r[0]:15s} | Score: {r[1]:20s} | Expansion: {r[2]:25s} | Time: {r[3]:.1f}s | R = {r[4]:.2f}")
results['Score'] = [record for record in records if record[2]=='ScoreBasedExpansion']
results['APPR'] = [record for record in records if record[2]=='APPR']

print(len(results['Score']),len(results['APPR']))
```

```python
rows = []
for x in results['SEDA']:
    rows.append({'dataset':x[0],
                'R':x[1],
                'time':x[2],
                "model":"SEDA"})

for x in results['Alatawi']:
    rows.append({'dataset':x[0],
                'R':x[1],
                'time':x[2],
                "model":"alatawi"})

for x in results['Score']:
    rows.append({'dataset':x[0],
                'R':x[4],
                'time':x[3],
                "model":"Score",
                'seed':x[1]})

for x in results['APPR']:
    rows.append({'dataset':x[0],
                'R':x[4],
                'time':x[3],
                "model":"APPR",
                'seed':x[1]})
```

```python
for dataset in df['dataset'].unique():
    subset = df[df['dataset'] == dataset].copy()

    # Convert R to float, keeping 'inf' as inf for conditional coloring
    subset['R'] = subset['R'].replace('inf', float('inf')).astype(float)

    fig, ax1 = plt.subplots(figsize=(10, 5))
    x_labels = [f"{m}\n{s}" if pd.notna(s) else m for m, s in zip(subset['model'], subset['seed'])]
    x = np.arange(len(subset))

    # Create colors and heights arrays
    colors = []
    heights = []
    for r in subset['R']:
        if r == float('inf'):
            colors.append('pink')
            heights.append(ax1.get_ylim()[1] if ax1.get_ylim()[1] > 0 else 2)  # Full height
        else:
            colors.append('skyblue')
            heights.append(r)

    # Plot R as bars with conditional coloring and heights
    bars = ax1.bar(x, heights, color=colors, alpha=0.6, label='R value')
    ax1.set_ylabel("R value", color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Plot time as line (secondary axis)
    ax2 = ax1.twinx()
    ax2.plot(x, subset['time'], color='red', marker='o', label='Time')
    ax2.set_ylabel("Time (s)", color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    k = 1
    ax2.axhline(y=k, color='black', linestyle=':', linewidth=1, label=f'k = {k}')

    # Formatting
    ax1.set_title(dataset)
    ax1.set_xticks(x)
    ax1.set_xticklabels(x_labels, rotation=75, ha='right')

    plt.tight_layout()
    plt.show()


```
